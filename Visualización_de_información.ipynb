# Visualización de información

En este caso usaremos las tablas de información sobre las 10000 empresas más grandes de Colombia de acuerdo con sus ingresos operativos [SIIS](https://siis.ia.supersociedades.gov.co/10MilEmpresas/index.html#/)

```python
# Cargamos las tablas mencionados
import pandas as pd

reporte2021 = pd.read_excel("/content/reporte10000_2021.xlsx")
reporte2022 = pd.read_excel("/content/reporte10000_2022.xlsx")
```

```python
reporte2021.head()
```

```python
reporte2022.head()
```

```python
estados2020 = pd.concat([reporte2021.iloc[:,0:9],reporte2021.iloc[:,14:20]], axis = 1)
estados2021 = reporte2021.iloc[:,0:14]
estados2022 = reporte2022.iloc[:,0:14]
```

```python
estados2020 = estados2020.drop("NIT", axis = 1)
```

```python
estados2021 = estados2021.drop("NIT", axis = 1)
estados2022 = estados2022.drop("NIT", axis = 1)
```

```python
estados2020.describe()
```

```python
estados2021.describe()
```

```python
estados2022.describe()
```

```python
index_empresa2020 = estados2020["GANANCIA (PÉRDIDA) 2020"].argmax()
empresa2020 = estados2020.iloc[index_empresa2020]["RAZÓN SOCIAL"]

index_empresa2021 = estados2021["GANANCIA (PÉRDIDA) 2021"].argmax()
empresa2021 = estados2021.iloc[index_empresa2021]["RAZÓN SOCIAL"]

index_empresa2022 = estados2022["GANANCIA (PÉRDIDA) 2022"].argmax()
empresa2022 = estados2022.iloc[index_empresa2022]["RAZÓN SOCIAL"]
```

```python
print("La empresa con la mayor ganancia para el año 2020 fue", empresa2020)
print("La empresa con la mayor ganancia para el año 2021 fue", empresa2021)
print("La empresa con la mayor ganancia para el año 2022 fue", empresa2022)
```

```python
#macrosector que más se repite en cada uno de los años
print("El macrosector con mayor prevalencia para el año 2020 fue", estados2020['MACROSECTOR'].mode()[0])
print("El macrosector con mayor prevalencia para el año 2021 fue", estados2021['MACROSECTOR'].mode()[0])
print("El macrosector con mayor prevalencia para el año 2022 fue", estados2022['MACROSECTOR'].mode()[0])
```

```python
#de forma gráfica
import seaborn as sns
sns.set()

sns.countplot(estados2020['MACROSECTOR'], color='blue');
```

```python

sns.countplot(estados2021['MACROSECTOR'], color='blue');
```

```python

sns.countplot(estados2022['MACROSECTOR'], color='blue');
```

```python
estados2020.groupby(["MACROSECTOR"])["GANANCIA (PÉRDIDA) 2020"].sum().plot(kind = 'bar');
```

```python
estados2021.groupby(["MACROSECTOR"])["GANANCIA (PÉRDIDA) 2021"].sum().plot(kind = 'bar');
```

```python
estados2022.groupby(["MACROSECTOR"])["GANANCIA (PÉRDIDA) 2022"].sum().plot(kind = 'bar');
```

```python
from matplotlib import cm
import matplotlib.pyplot as plt

cmap = [100,200,300,400,500,1]

estados2020.groupby('MACROSECTOR')[["GANANCIA (PÉRDIDA) 2020",
                                   "INGRESOS OPERACIONALES 2020"]].sum().plot.scatter(x = "INGRESOS OPERACIONALES 2020"
                                    , y='GANANCIA (PÉRDIDA) 2020', c = cmap);


plt.legend(['AGROPECUARIO','COMERCIO'], loc = "best", title='Macrosector')
plt.show()
```
